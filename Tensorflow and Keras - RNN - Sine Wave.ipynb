```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate data X

x = np.linspace(0,50,501) 
#np.linspace = 0 sampai 50 dengan 501 point
#np.arange / range = 0 sampai 50 dengan delta sekian
x
```

```python
# generate daya y dengan rumus sin(x)
y = np.sin(x)
y
```

```python
plt.plot (x,y)
```

```python
#dibuat ke dalam dataframe
df = pd.DataFrame(data=y, index=x, columns=['Sine'])
df
```

```python
#Train-Test Split

# splitnya bukan random, tapi mundur ke belakang
# berapa persen data test dari t samapai t-n?
# kl dilihat index dari index 50 sampai berapa?
print('jumlah data: ')
print(len (df))
test_percent = 0.1
len(df)*test_percent
```

```python
#pembulatan
test_point = np.round(len(df)*test_percent)
test_point
#print (type(test_point))
```

```python
#test index value 
#->untuk menentukna cut-off/ titik dijadikan patokan 
test_ind = int(len(df)-test_point)
#can be used for indexing with pandas
test_ind
#print (type(test_ind))
```

```python
train = df.iloc[:test_ind] #dari row awal - test_ind
test = df.iloc[test_ind:] #dari row test_ind - akhir
train
```

```python
test
```

```python
#scale the Y label
# karena RNN cara kerjanya outputnya akan di feed ulang ke neuron menjadi input
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(train) 
#sesuaikan scaler dengan data train
#FIT TETAP HANYA DI FEATURE SAJA
```

```python
#transform data train dan test
scaled_train = scaler.transform(train)
scaled_test = scaler.transform(test)
scaled_train
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
#help(TimeseriesGenerator)
length = 25 #panjang sekuens outpunya
batch_size = 1 #jumlah sample dalam 1 batch

generator = TimeseriesGenerator(scaled_train, scaled_train,
                                length=length, batch_size=batch_size)
#(X,y label)
#scaled trainin data has both X and y
```

```python
length = 50 #panjang sekuens outpunya
batch_size = 1 #jumlah sample dalam 1 batch

generator = TimeseriesGenerator(scaled_train, scaled_train,
                                length=length, batch_size=batch_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN,LSTM
```

```python
n_feature = 1
model = Sequential()
model.add(SimpleRNN(50, input_shape=(length,n_feature)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
model.fit_generator(generator,epochs=5)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```
